```python
import pandas as pd
import numpy as np
from sklearn import linear_model # import linear regression library from matplot lib
import matplotlib.pyplot as plt
```

```python
# define training set that consist of three points (assume 1d only)
X_train = [[1,], [2,], [5,]]
Y_train = [3, 6, 8]
```

```python
# Baysian Ridge linear regression
clf = linear_model.BayesianRidge()
clf.fit(X_train, Y_train)
```

```python
clf.predict([[3,]], return_std=True)
```

```python
# plot the date
X_data=[1,2,5]
Y_data=[3,6,8]
plt.scatter(X_data, Y_data, marker='x')
x_input = np.linspace(0,6,100)
y_input = np.zeros_like(x_input)
y_input_minus = np.zeros_like(x_input)
y_input_plus = np.zeros_like(x_input)
for index, x in enumerate(x_input):
    pred, std = clf.predict([[x,]], return_std=True)
    y_input[index] = pred[0]
    y_input_minus[index] = pred[0] - std[0]
    y_input_plus[index] = pred[0] + std[0]

plt.plot(x_input, y_input, '-')
plt.plot(x_input, y_input_minus, '--')
plt.plot(x_input, y_input_plus, '--');
```
